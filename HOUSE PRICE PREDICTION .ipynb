# Spliting the data 

#### Without Scalling 

```python
#spliting the data
x=df.drop(['price','id','date'], axis=1)
```

```python
x
```

```python
y=df['price']
```

```python
y
```

```python
# spliting the data
from sklearn.model_selection import train_test_split

# split the data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

```

```python
x_train
```

```python
y_train
```

```python
y_test
```

# MODEL FITING AND ACCURACY CHECK 

```python
# importing libraries
from sklearn.linear_model import LinearRegression

# create a model
lr=LinearRegression()

# fiting the model
lr.fit(x_train,y_train)

# predict the model
y_pred=lr.predict(x_test)

print('y_pred : {}'.format(y_pred))

print('  ')

# Score Checkin of Our Model 
# To Check Accuracy of Linear Regression model without scalling

from sklearn.metrics import *

lr_score=lr.score(x_test,y_test)

lr_score

print('Linear Regression Model Accuracy : {}'.format(lr_score))
```

```python
# WITH SCALLING  
# with scalling the data

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

sc.fit(x_train)

x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)


lr.fit(x_train_sc,y_train)
y_pred_sc=lr.predict(x_test_sc)
scale_accuracy=lr.score(x_test_sc,y_test)

print('Scalling Data Accuracy : {}'.format(scale_accuracy))
```

# MODEL EVALUATION 

```python
# cost function of regression
# to check model evaluation of linear regression 
mse=mean_squared_error(y_test,y_pred)
mean_score=mean_squared_error(y_pred,y_test)
# check error
rmse=np.sqrt(mean_score)

print('Mean Square Error : {}'.format(mse))
print('Root Mean Square Error : {}'.format(rmse))

# find out R Square value find out 

r2_score=r2_score(y_pred,y_test)
print('r2_score : {}'.format(r2_score))
```

#### Now Implementing a ridge and lasso regularization (L1 & L2 regularization)

```python
# implimenting ridge and lasso regularization L1 & L2 REGULARIZAION

from sklearn.linear_model import Ridge,Lasso


# Ridge Regulariaztion
# fitting the model

ridge=Ridge(alpha=2)

# fit the model

ridge.fit(x_train,y_train)
ridge_score=ridge.score(x_test,y_test)

print('Ridge regularizatio model accuracy : {}'.format(ridge_score))
# Lasso Regularization 


lasso=Lasso(alpha=6)
lasso.fit(x_train,y_train)
lass_score=lasso.score(x_test,y_test)

print('Lasso Regularization model accuracy : {}'.format(lass_score))
```

# POLYNOMINAL REGRESSION 

```python
# Polynominal regression
# Polynomial Regression

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly_reg=PolynomialFeatures(degree=2)

# fit the model
poly_reg.fit(x_train)
x_train_poly=poly_reg.transform(x_train)
x_test_poly=poly_reg.transform(x_test)

# use first linear regression
lr=LinearRegression()

lr.fit(x_train_poly,y_train)


polynonimal_score=lr.score(x_test_poly,y_test)

print('Polynominal model score : {}'.format(polynonimal_score))



```
